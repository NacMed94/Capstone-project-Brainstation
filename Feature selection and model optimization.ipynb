## Re-merging discharge classes 

```python
# Reading dataframes
training_v1 = pd.read_pickle('G:/My Drive/Capstone-Data/pickles/training_df_v1.pkl')
X_tst_v1 = pd.read_pickle('G:/My Drive/Capstone-Data/pickles/X_test_v1.pkl')
events_tst_v1 = pd.read_pickle('G:/My Drive/Capstone-Data/pickles/y_test_v1.pkl')
```

Let's extract the features and try a simple logistic regression model. The aim now is to see how models react to the discharge events. If we see a tendency to group them together, we will, as this will help us decrease the number of classes. We are doing this because we are not interested in predicting discharge type, rather transfer type (and discharge vs transfer). We can binarise our y datasets using sklearn's LabelEncoder.

```python
from sklearn.preprocessing import LabelEncoder

# Only feature columns
X_trn_v1 = training_v1.drop(columns = ['event','eventtype'])

# Event column. Label encoding for the fitting (strings converted to integers from 1 to length of unique events)
le_v1 = LabelEncoder()
le_v1.fit(training_v1.event)
y_trn_v1 = le_v1.transform(training_v1.event)
y_tst_v1 = le_v1.transform(events_tst_v1)
```

```python
# Importing functions and classes
from sklearn.preprocessing import MinMaxScaler  
from sklearn.preprocessing import StandardScaler  
from sklearn.linear_model import LogisticRegression 
from sklearn.pipeline import Pipeline 
from sklearn.model_selection import GridSearchCV # Grid search tool to try multiple parameters and models
from sklearn.decomposition import PCA

cv = 5 

# Defining pipeline
# In our pipeline, the data will be normalised before a logistic regression is fitted 
estimators = [('scaler', StandardScaler()),
              ('model',LogisticRegression())] # And logistic regression only (as specified in grid as well)

pipe = Pipeline(estimators)
# Defining parameter grid

C_range = np.logspace(1,10,num = 4) # Setting range of C values

param_grid = [
    {
        'scaler':[StandardScaler()], # Normalisations
        'model':[LogisticRegression()], # Model
        'model__C':C_range, # Regularisation parameter
    }
]

# Creating grid using 10 folds (this is the standard and we have enough data so that the training set will not be too small)
grid_logreg = GridSearchCV(pipe, param_grid, cv = cv)
# Fitting grid (or models)
logreg_v1_1 = grid_logreg.fit(X_trn_v1, y_trn_v1)
display(logreg_v1_1.best_estimator_)
print(logreg_v1_1.best_score_)

dump(logreg_v1_1,G:/My Drive/Capstone-Data/models/logreg_v1_1.joblib')
```

## Random forest

Let's try a random forest so that we can do a feature selection and fit a model. Random forest could work better than logistic regression as it is likely that there are few linear relationships between our features and event (based on correlation of EDA) and that feature selection via PCA does not seem to be very relevant. 

Let's start by tuning the hyperparameters. We can try with a relatively simple model setting the min samples split to be very high (so that nodes are only split if we have a certain number of samples). This will decrease the complexity of the tree, so the models can be fitted relatively fast and we may have less overfitting as we see in the previous random forest models.

Note that we do not need to use a scaler, as random forest are quite insensitive to data scaling.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

cv = 5 

pipe = Pipeline([('model',RandomForestClassifier())])

param_grid = [
    {
        'model':[RandomForestClassifier()], # Model
        'model__min_samples_split':[0.01, 0.05, 0.1], # Min samples split is a ratio of the samples
    }
]

grid = GridSearchCV(pipe, param_grid, cv = cv)

rfccv_v2_3_A = grid.fit(X_trn_v2, y_trn_v3)
display(rfccv_v2_3_A.best_estimator_)

nachopy.roc_n_confusion(rfccv_v2_3_A,
                        X_trn_v2,
                        y_trn_v3,
                        lbl_enc=le_v3
                       )
dump(rfccv_v2_3_A,r'G:/My Drive/Capstone-data/models/rfccv_v2_3_A.joblib');
```

That was not great, but better in some aspects than the logistic regression models. We see a decrease in average score but the AUC has increase to 0.93. We predict well the majoritary class (Discharge) with a recall of 1, although many patients that ended up in other care units have been predicted as if they were being discharged, particularly for Neuro, Pacu, transpland and Trauma. The recall for these care units was 0 in fact, like for MICU/SICU and ED Observation. Discharge Lounge and Gyn/Obs/Labour care units are often predicted correctly because they have more unique sets of associated features, as seen during the EDA.

To make better predictions, we may need to make our models more complex. We will fit another model tuning min_samples_leaf (minimum amount of samples in terminal node), which seems easier to control than the min_samples_split. We will set these to be low to allow the tree to form terminal nodes with few samples, which may be favourable for our minority classes, which have much fewer samples than discharge and Medicine. This will however be limited by the max_depth (the number of partitions) which will be have a maximum of 80. To avoid spending days fitting future models, we will find the minimum number of estimators that work best at different depths. The higher the number of estimators, the lower the variance of the model, so less prone to overfitting. This is because we are using the outputs of many different models for the prediction. Ideally, we would find the minimum number of estimators that decreases overfitting as once this point is reached we see minimal decreases in variance but the computation time keeps increasing. 

```python
cv = 5 

pipe = Pipeline([('model',RandomForestClassifier())])

param_grid = [
    {
        'model':[RandomForestClassifier()], # Most default hyperparameters are kept. The criterion is left to gini for speed
        'model__n_estimators': [33,66,100], # Th
        'model__max_depth':[20, 40, 80], 
        'model__min_samples_leaf':[10,100,1000] 
    }
]

grid = GridSearchCV(pipe, param_grid, cv = cv)

rfccv_v2_3_B = grid.fit(X_trn_v2, y_trn_v3)
display(rfccv_v2_3_B.best_estimator_)

nachopy.roc_n_confusion(rfccv_v2_3_B,
                        X_trn_v2,
                        y_trn_v3,
                        lbl_enc=le_v3
                       )
dump(rfccv_v2_3_B,r'G:/My Drive/Capstone-data/models/rfccv_v2_3_B.joblib');
```

Much better, although predicting correctly minoritary classes seems really hard. The best model had 100 estimators, a minimum leaf (terminal node) of 10 and maximum depth of 40. . By setting a limit to the number of samples we can have on the terminal nodes, we are not letting the trees "ramify" fine enough to find the right relationships between the data for these classes. We can avoid these hyperparameters for now and focus on finding the right combination of max_depth with other hyperparameters, such as max_features. Max_features is the maximum number of features considered at each node split. The default is the square root of the number of features, let's try somehting lower (the log2) and higher (20% of the number of features)

It would be ideal if we could tune the feature selection as if it were a hyperparameter, like with the PCA. For this, we can create a custom class that applies the transformation just by selecting as features for the model the top x% most important columns

```python
# Needed placeholders for sklearn transformer classes
from sklearn.base import BaseEstimator, TransformerMixin

class FeatureSelector(BaseEstimator, TransformerMixin):
    # Transformer to select top features (columns must be sorted by order of importance)
    # n_features is a float to indicate the features to use
    def __init__(self, n_features = 1):
        self.n_features = n_features
        
    def fit(self, X, y = None):
        return self
    
    def transform(self, X, y = None):
        # The last column is selected 
        last_col = round(self.n_features * len(X.columns))
        X_ = X[X.columns[:last_col]].copy()
        
        return X_
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

cv = 5 

pipe = Pipeline([('transformer',FeatureSelector()),
                 ('model',RandomForestClassifier())])

param_grid = [
    {
        'transformer':[FeatureSelector()],
        'transformer__n_features':[0.1,0.25,0.6,1],
        'model':[RandomForestClassifier()], # Model
        'model__max_depth':[40],
        'model__max_features':['sqrt','log2', 0.2],
        'model__min_samples_leaf':[2, 5, 10],
        'model__n_jobs':[-1]  # Using all CPUs
    }
]

grid = GridSearchCV(pipe, param_grid, cv = cv, verbose = 1)

rfccv_v2_3_C = grid.fit(X_trn_v2, y_trn_v3)
display(rfccv_v2_3_C.best_estimator_)

nachopy.roc_n_confusion(rfccv_v2_3_C,
                        X_trn_v2,
                        y_trn_v3,
                        lbl_enc=le_v3
                       )
dump(rfccv_v2_3_C,'G:/My Drive/Capstone-data/models/rfccv_2_3_C.joblib');
```

In terms of hyperparameters, we see that the best model was the most complex one: the lowest minimum sample leaf and the maximum max_features, and n_features (from FeatureSelector, so all features from matrix were added to the model). Let's create another y vector from the eventtype of training_df (merged discharge and transfer classes) to see if we can adapt our model to a binary classification problem.

We will do a relatively simple grid search. Max features will be kept at 0.2, which worked well for previous models and max_depth will be adjusted between 20 and 80. For the binary problem we may not need all the features, so we are also including the FeatureSelector transformer to select and use only the top features to fit the model.

```python
y_trn_v4 = training_v2_3.eventtype
y_tst_v4 = (events_tst_v3 == 'Discharge')*1

cv = 5 

pipe = Pipeline([('transformer',FeatureSelector()),
                 ('model',RandomForestClassifier())])

param_grid = [
    {
        'transformer':[FeatureSelector()],
        'transformer__n_features':[0.5, 0.7, 0.9, 1], # Trying feature selection again
        'model':[RandomForestClassifier()],
        'model__max_depth':[20,40,80],
        'model__max_features':[0.2],
        'model__n_jobs':[-1]  # Using all CPUs
    }
]

grid = GridSearchCV(pipe, param_grid, cv = cv, verbose = 2)

rfccv_v2_4 = grid.fit(X_trn_v2, y_trn_v4)
display(rfccv_v2_4.best_estimator_)

nachopy.roc_n_confusion(rfccv_v2_4,
                        X_trn_v2,
                        y_trn_v4,
                       )
dump(rfccv_v2_4,'G:/My Drive/Capstone-data/models/rfccv_2_4_C.joblib');
# Please, ignore half-plotted graph in the bottom.
```
